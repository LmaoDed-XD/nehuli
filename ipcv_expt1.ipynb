```python
#Arithmetic Operations
```

```python
image1 = cv2.imread('sea_link.jpg')
```

```python
image2 = cv2.imread('gateway.jpg')
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
```

```python
print(image1.shape)
```

```python
print(image2.shape)
```

```python
image2 = cv2.resize(image2, (544,360))
```

```python
plt.imshow(image2)
```

```python
plt.imshow(image1)
```

```python
#WeightedSum
```

```python
weightedSum = cv2.addWeighted(image1, 0.5, image2, 0.4, 0)
plt.imshow(weightedSum)
```

```python
#Subtraction
```

```python
subtraction = cv2.subtract(image1, image2)
plt.imshow(subtraction)
```

```python
subtraction1 = cv2.subtract(image2, image1)
plt.imshow(subtraction1)
```

```python
#Bitwise Opertations
```

```python
img_and = cv2.bitwise_and(image1, image2)
```

```python
plt.imshow(img_and)
```

```python
img_or = cv2.bitwise_or(image1, image2)
plt.imshow(img_or)
```

```python
img_xor = cv2.bitwise_xor(image1, image2)
plt.imshow(img_xor)
```

```python
img_not = cv2.bitwise_not(image1, image2)
plt.imshow(img_not)
```
