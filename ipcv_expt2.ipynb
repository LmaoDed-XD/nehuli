```python
import cv2, matplotlib.pyplot as plt, numpy as np
```

```python
image = cv2.imread('gateway.jpg')
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
gray = cv2.imread('gateway.jpg',0)
```

```python
color_negative = abs(255-img)
```

```python
x,y = gray.shape
gray_negative = np.zeros((x,y))
for i in range (0,x):
    for j in range (0,y):
        gray_negative[i][j] = abs(255 - gray[i][j])      
```

```python
imgs = [img, gray, color_negative, gray_negative]
title = ['colored', 'gray', 'colored_negative', 'gray_negative']

plt.subplot(2,2,1)
plt.imshow(imgs[0])
plt.title(imgs[0])
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,2)
plt.imshow(imgs[1], cmap = 'gray')
plt.title(title[1])
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,3)
plt.imshow(imgs[2])
plt.title(title[2])
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,4)
plt.imshow(imgs[3], cmap = 'gray')
plt.title(title[3])
plt.xticks([])
plt.yticks([])

plt.show()
```

```python
#thresholding
```

```python
image = cv2.imread('gateway.jpg', 0)
x,y = image.shape
z = np.zeros((x,y))
for i in range(0,x):
    for j in range(0,y):
        if (image[i][j]>50):
            z[i][j]=255
        else:
            z[i][j]=0
equ = np.hstack((image,z))
plt.title("Thresholding")
plt.imshow(equ,'gray')
plt.show()
```

```python
#graylevel slicing
```

```python
image = cv2.imread('gateway.jpg', 0)
x,y = image.shape
z = np.zeros((x,y))
for i in range(0,x):
    for j in range(0,y):
        if (image[i][j]>50 and image[i][j]<150):
            z[i][j]=255
        else:
            z[i][j]=image[i][j]
equ = np.hstack((image,z))
plt.title("Graylevel slicing with background")
plt.imshow(equ,'gray')
plt.show()
```
